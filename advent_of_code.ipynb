```python
def range_overlap(range1, range2):
    """Whether range1 and range2 overlap fully."""
#     x1, x2 = range1.start, range1.stop
#     y1, y2 = range2.start, range2.stop
    xs = set(range1)
    xs = xs.intersection(range2)
    if xs:
        return True
    
    #return (x1 <= y1 and x2 >= y2) or (y1 <= x1 and y2 >= x2) 
```

```python
range_overlap(range(2,7),range(1,6))
```

```python
for i in lines:
    first_start = int(str.split(i,',')[0].split('-')[0])
    first_end = int(str.split(i,',')[0].split('-')[1])
    
    second_start = int(str.split(i,',')[1].split('-')[0])
    second_end = int(str.split(i,',')[1].split('-')[1])
    
    if range_overlap(range(first_start,first_end+1),range(second_start,second_end+1)):
        count+=1
```
